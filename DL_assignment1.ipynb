```python
import numpy as np
import torch
```

## Different Tensors in numpy


```python
# 1D
a_np = np.array([5,7,3,2,1,4])

# 2D
b_np = np.array([[4, 3, 5, 3, 5],
                 [8, 2, 6, 2, 3]])

# 3D
c_np = np.array([[[1, 5, 5, 5], [5, 2, 3, 4]],
                 [[9, 3, 6, 2], [7, 8, 2, 4]]])

print(a_np.shape, b_np.shape, c_np.shape)
```

## Tensors in Pytorch

```python
# 1D
a_t = torch.tensor([5,7,3,2,1,4])

# 2D
b_t = torch.tensor([[4, 3, 5, 3, 5],
                    [8, 2, 6, 2, 3]])

# 3D
c_t = torch.tensor([[[1, 5, 5, 5], [5, 2, 3, 4]],
                    [[9, 3, 6, 2], [7, 8, 2, 4]]])

print(a_t.shape, b_t.shape, c_t.shape)
```

##Basic Operations: Element-wise addition, subtraction, multiplication, and division.

### using pytorch

```python
x = torch.tensor([5, 2, 4])
y = torch.tensor([4, 8, 4])

add = torch.add(x, y)
sub = torch.sub(x, y)
mul = torch.mul(x, y)
div = torch.div(x, y)

print(add, sub, mul, div)
```

### using numpy

```python
x_np = np.array([5, 2, 4])
y_np = np.array([4, 8, 4])

print(x_np + y_np)
print(x_np - y_np)
print(x_np * y_np)
print(x_np / y_np)
```

## DOT product

### using pytorch


```python
x = torch.tensor([5, 2, 4])
y = torch.tensor([4, 8, 4])

dot = torch.dot(x, y)
print(dot)
```

### using numpy


```python
x_np = np.array([5, 2, 4])
y_np = np.array([4, 8, 4])

np.dot(x_np, y_np)
```

## Matrix Multiplication

### using pytorch

```python
A = torch.tensor([[5, 7],
                  [4, 8]])
B = torch.tensor([[2, 7],
                  [3, 5]])

matmul1 = torch.matmul(A, B)
#OR
matmul2 = A @ B

print(matmul1)
print(matmul2)
```

### using numpy

```python
x_np = np.array([[4, 3],
                 [8, 2]])
y_np = np.array([[1, 5],
                 [5, 2]])

np.matmul(x_np, y_np)
```

## slicing, extracting subtensors using pytorch

```python
t = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print(t[0])
print(t[:, 1])
print(t[1, 2])
```

## slicing, extracting subtensors using pytorch

```python
t = np.array([[10, 20, 30],
                  [40, 50, 60]])

print(t[0])
print(t[:, 1])
print(t[1, 2])
```

## Boolean masking using pytorch

```python
t = torch.tensor([1, 2, 3, 4, 5])

mask = t > 3
filtered = t[mask]

print(filtered)
```

## Boolean masking using numpy

```python
t = np.array([1, 2, 3, 4, 5])
t[t > 3]
```

## view in pytorch

```python
x = torch.tensor([1, 2, 3, 4, 5, 6])

x_view = x.view(2, 3)
print(x_view)
```

## reshape in pytorch

```python
x_reshape = x.reshape(3, 2)
print(x_reshape)
```

## squeeze in pytorch

```python
x = torch.tensor([[1, 2, 3]])

print(x.shape)
print(x.squeeze().shape)
```

## unsqueeze in pytorch

```python
x = torch.tensor([1, 2, 3])

print(x.shape)
print(x.unsqueeze(0).shape)
print(x.unsqueeze(1).shape)
```

## reshape in numpy

```python
x_np =np.array([1, 2, 3, 4, 5, 6])
x_np.reshape(2, 3)
```

## Brodcasting in pytorch

```python

x = torch.tensor([[5, 2, 8],
                  [3, 5, 2]])

y = torch.tensor([10, 90, 30])

result = x + y
print(result)
```

## Brodcasting in numpy

```python
x_np = np.array([[5, 2, 8],
                  [3, 5, 2]])
y_np = np.array([10, 90, 30])

print(x_np + y_np)
```

## out-of-place operation in pytorch


```python
x = torch.tensor([5, 7, 8])
y = x + 1

print(x)
print(y)
```

## in-place opearion in pytorch

```python
x = torch.tensor([6, 3, 5])
x.add_(1)

print(x)
```

##out-of-place operation in numpy

```python
arr = np.array([1, 2, 3, 4, 5])
arr_y = arr + 10

print(arr)
print(arr_y)
```

## in-place operation in numpy


```python
arr = np.array([1, 2, 3, 4, 5])
arr += 10
print(arr)
```
